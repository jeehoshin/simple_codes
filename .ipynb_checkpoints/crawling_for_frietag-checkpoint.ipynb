```python
from selenium import webdriver
import schedule
import time
from selenium.webdriver.common.keys import Keys
import time
import pyperclip
```

```python
print('10분마다 프라이탁 홈페이지를 갱신하며, 신제품 여부를 확인합니다.\n')
driver = webdriver.Chrome('./chromedriver')
 
preLinks = []
newLinks = []
 

def writeComment():
    global preLinks
    global newLinks
    driver = webdriver.Chrome('./chromedriver')
    url = 'https://www.naver.com/'
    driver.get(url)
    # 로그인 버튼을 찾고 클릭합니다
    login_btn = driver.find_element_by_class_name('link_login')
    login_btn.click()
    time.sleep(3)
    # id, pw 입력할 곳을 찾습니다.
    tag_id = driver.find_element_by_name('id')
    tag_pw = driver.find_element_by_name('pw')
    tag_id.clear()
    time.sleep(1)
    # id 입력
    tag_id.click()
    pyperclip.copy('daleji21')
    tag_id.send_keys(Keys.CONTROL, 'v')
    time.sleep(1)
    
    # pw 입력
    tag_pw.click()
    pyperclip.copy('fpemgkt8413')
    tag_pw.send_keys(Keys.CONTROL, 'v')
    time.sleep(1)
    
    # 로그인 버튼을 클릭합니다
    login_btn = driver.find_element_by_id('log.login')
    login_btn.click()
    
    #메일버튼을 클릭합니다.
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="NM_FAVORITE"]/div[1]/ul[1]/li[1]/a').click()
    
    #메일쓰기를 클릭합니다 
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="nav_snb"]/div[1]/a[1]').click()
    
    #받는사람을 지정합니다 
    time.sleep(1)
    driver.find_element_by_css_selector("textarea#toInput").send_keys("sjiho0747@gmail.com")
    
    #제목을 써줍니다
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="subject"]').send_keys('하파오 입고!')
    
    #보냅니다!
    driver.find_element_by_xpath('//*[@id="sendBtn"]').click()
    preLinks = newLinks
    
    print(time.strftime('%c', time.localtime(time.time())) + ' ::: 새 제품 입고! 메일 발송했습니다')
    driver.close()
    
def getFreitag():
    print(time.strftime('%c', time.localtime(time.time())) + ' ::: 프라이탁에서 새로 불러옵니다.')
    global newLinks
    global preLinks
    url = 'https://www.freitag.ch/en/f41'
    driver.get(url)
    
 # 쿠키 노땡큐
    driver.find_element_by_css_selector('a#dismiss-cookies-cta').click()
    driver.find_element_by_css_selector('i.mdi.mdi-chevron-down').click()

    for item in driver.find_elements_by_css_selector('ul.products-list > li > a > img'):
        newLinks.append([item.get_attribute('src')])
 
    newItems = [x for x in newLinks if x not in preLinks]
 
 
    if len(newItems) != 0:
        writeComment()
    else:
        print("신제품이 없습니다.")

        
#첨에 일단 실행
getFreitag()
 
# 10분에 한번씩 실행
schedule.every(1).minutes.do(getFreitag)
 
while True:
    schedule.run_pending()
    time.sleep(1)
```
